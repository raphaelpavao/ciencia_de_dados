```python
class No: 
    def __init__(self, key): 
        self.left = None 
        self.right = None 
        self.val = key 

def inserir(root, key): 
    if root is not None:
        print(f"Comparando com nó: {root.val}, inserindo: {key}")
    else:
        print(f"Inserindo novo nó com valor: {key} (posição vazia)")

    if root is None: 
        return No(key) 
    else: 
        if root.val < key: 
            root.right = inserir(root.right, key) 
            print("Pegando a direita")
        else: 
            root.left = inserir(root.left, key) 
            print("Pegando a esquerda")
    return root 

def em_ordem(no):
    if no:
        em_ordem(no.left)
        print(no.val, end=" ")
        em_ordem(no.right)

def pre_ordem(no):
    if no:
        print(no.val, end=" ")
        pre_ordem(no.left)
        pre_ordem(no.right)

def pos_ordem(no):
    if no:
        pos_ordem(no.left)
        pos_ordem(no.right)
        print(no.val, end=" ")

def imprimir_arvore(no, nivel=0):
    if no is not None:
        imprimir_arvore(no.right, nivel + 1)
        print("    " * nivel + str(no.val))
        imprimir_arvore(no.left, nivel + 1)



raiz = No(14)# Raiz da árvore com o primeiro valor 
sequence = [4, 18, 0, 21, 17, 1, 8, 13]  

for key in sequence: 
    inserir(raiz, key) 




```
